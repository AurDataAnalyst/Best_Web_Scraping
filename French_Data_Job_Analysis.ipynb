```python
import requests
import bs4
import string
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
#import nltk
#from nltk.tokenize import word_tokenize
#from nltk.corpus import stopwords
#from nltk.tokenize import RegexpTokenizer
#from sklearn import preprocessing, linear_model, naive_bayes, metrics, svm
#from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
#city_set = ("Paris", "Nanterre")
#state_set = ("Texas", "California", "New York", "Florida", "North Carolina", "Colorado", "Washington")

state = "Hauts-de-Seine"

state_set = ("Hauts-de-Seine", "Paris")

job_title_set = ("business+intelligence", "data+analyst", "data+scientist", "data+engineer", "développeur+full+stack")

job_title = "developpeur+full+stack"

#job_title_set = ("business+intelligence", "data+analyst")

max_results_per = 10

columns = ["state", "full_job_title", "company_name", "url", "job_class"]
df = pd.DataFrame(columns = columns)
df.head()
```

```python
#scraping code:
for title in job_title_set:
    for state in state_set:
        print(title)
        print(state)

        for start in range(0, max_results_per):
            url = "https://www.indeed.fr/jobs?q=" + str(title) + "\"&l=" + str(state) + "&start=" + str(start)
            print(url)
            page = requests.get(url)
            time.sleep(1+random.random()*4)  #ensuring at least 1 second between page grabs
            soup = BeautifulSoup(page.text, "lxml", from_encoding="utf-8")

            for div in soup.find_all(name="div", attrs={"class":"row"}):
                #specifying row num for index of job posting in dataframe
                num = (len(df) + 1) 

                #creating an empty list to hold the data for each posting
                job_post = [] 

                #append state name
                job_post.append(state)

                #grabbing job title
                for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
                    job_post.append(a["title"]) 

                #grabbing company name
                company = div.find_all(name="span", attrs={"class":"company"}) 

                if len(company) > 0: 
                    for b in company:
                        job_post.append(b.text.strip()) 
                else: 
                    sec_try = div.find_all(name="span", attrs={"class":"result-link-source"})
                    for span in sec_try:
                        job_post.append(span.text)

            #grabbing location name
            #c = div.findAll('span', attrs={'class': 'location'})

            #for span in c:
                #job_post.append(span.text) 

            #grabbing summary text
            #d = div.findAll(name="a", attrs={"data-tn-element": "href"})

            #for span in d:
                #job_post.append(span.text.strip()) 

                #grabbing job url
                for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
                    job_post.append("http://www.indeed.fr" + a["href"]) 

                job_post.append(title)
            #grabbing salary
            #try:
                #job_post.append(div.find('nobr').text)
            #except:
                #try:
                    #div_two = div.find(name="div", attrs={"class":"sjcl"}) 
                    #div_three = div_two.find("div") 
                    #job_post.append(div_three.text.strip())
                #except:
                    #job_post.append("Nothing_found") 

                print(job_post)
                #appending list of job post info to dataframe at index num
                df.loc[num] = job_post
    
#saving df as a local csv file — define your own local path to save contents 
df.to_csv("outputfullstack.csv", encoding='utf-8')
```
